```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Reorder Collumn Headers
school_data_complete = school_data_complete.rename(columns = {'student_name': 'Student Name', 'school_name':'School Name', 
                        'reading_score':'Reading Score', 'maths_score':'Maths Score', 'type':'School Type', 'size':'Total Students', 'budget': 'Total School Budget'})
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = (school_data_complete['School Name'].nunique())

# Calculate the total number of students
total_students = int(school_data_complete['Student ID'].nunique())

# Calculate the total budget
total_budget = school_data_complete['Total School Budget'].unique().sum()
#STRtotal_budget = str(total_budget)
#total_budget = "$" + STRtotal_budget

# Calculate the average maths score 
total_avemath = round(school_data_complete['Maths Score'].mean(), 2)

# Calculate the average reading score
total_averead = round(school_data_complete['Reading Score'].mean(), 2)

# Calculate the percentage of students with a passing maths score (50 or greater)
mathpass_df = school_data_complete.loc[ school_data_complete["Maths Score"] >= 50,:]
mathpass_count = mathpass_df["Student ID"].count()
mathpass_percent = round((mathpass_count/total_students)*100,2)

# Calculate the percentage of students with a passing reading score (50 or greater)
readpass_df = school_data_complete.loc[ school_data_complete["Reading Score"] >= 50,:]
readpass_count = readpass_df["Student ID"].count()
readpass_percent = round((readpass_count/total_students)*100,2)

# Calculate the percentage of students who passed maths **and** reading (% Overall Passing)
mathreadpass_df = school_data_complete.loc[(school_data_complete["Reading Score"] >= 50) & 
                                      (school_data_complete["Maths Score"] >= 50),:]
mathread_count = mathreadpass_df["Student ID"].count()
overallpass_percent = round((mathread_count/total_students)*100,2)

# Create a dataframe to hold the above results
# Local Government Area Summary (LGAS)

LGAS = pd.DataFrame({'Total Schools':[total_schools],'Total Students': [total_students],
                      'Total Budget':[total_budget], '% Average Maths Score':[total_avemath],
                      '% Average Reading Score':[total_averead], '% Passing Maths':[mathpass_percent],
                       '% Passing Reading':[readpass_percent], '% Overall Passing':[overallpass_percent]})

# Use Map to format all the columns
LGAS["Total Students"] = LGAS["Total Students"].astype(float).map("{:,.0f}".format)
LGAS["Total Budget"] = LGAS["Total Budget"].astype(float).map("${:,.0f}".format)

# Optional: give the displayed data cleaner formatting
LGAS.head()

# export LGAS to csv
#LGAS.to_csv('LGAS.csv', index=False)
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Average Maths Score
average_maths = pd.DataFrame(school_data_complete.groupby(["School Name"])["Maths Score"].mean())
average_maths = average_maths.rename(columns = {"Maths Score":"Average Maths Score"})
average_maths = average_maths[["Average Maths Score"]].sort_values("Average Maths Score", ascending=False)

# Average Reading Score
average_reading = pd.DataFrame(school_data_complete.groupby(["School Name"])["Reading Score"].mean())
average_reading["Average Reading Score"] = average_reading["Reading Score"]
average_reading = average_reading[["Average Reading Score"]].sort_values("Average Reading Score", ascending=False)

# % Passing Maths
passing_maths = pd.DataFrame(school_data_complete.loc[school_data_complete["Maths Score"] >= 50,:])
passing_maths = pd.DataFrame(passing_maths.groupby("School Name").count())
passing_maths = passing_maths.rename(columns = {"Maths Score":"% Passing Maths"})
passing_maths = passing_maths[["% Passing Maths"]].sort_values("% Passing Maths", ascending=False)

# % Passing Reading
passing_reading = pd.DataFrame(school_data_complete.loc[school_data_complete["Reading Score"] >= 50,:])
passing_reading = pd.DataFrame(passing_reading.groupby("School Name").count())
passing_reading = passing_reading.rename(columns = {"Reading Score":"% Passing Reading"})
passing_reading = pd.DataFrame(passing_reading[["% Passing Reading"]]).sort_values("% Passing Reading", ascending=False)

# % Overall Passing (The percentage of students that passed maths and reading.)
overall_passing = pd.DataFrame(school_data_complete.loc[(school_data_complete["Reading Score"] >= 50)
                                                & (school_data_complete["Maths Score"] >= 50),:])
overall_passing  = pd.DataFrame(overall_passing.groupby("School Name").count())
overall_passing = overall_passing.rename(columns = {"Student ID":"% Overall Passing"})
overall_passing = overall_passing[["% Overall Passing"]].sort_values("% Overall Passing", ascending=False)
```

```python
#Compiling School Summary Table - drop duplicate rows

# School Name, School Type, Total Students
school_summary = school_data_complete[["School Name","School Type",
                                "Total Students","Total School Budget"]
                              ].drop_duplicates(["School Name"])

# Total School Budget, Per Student Budget
school_summary["Per Student Budget"] = school_summary["Total School Budget"]/school_summary["Total Students"]
school_summary.head()
```

```python
#Compiling the data into a single dataset.  
school_summary = school_summary.merge(average_maths,on='School Name').merge(average_reading,on='School Name').merge(passing_maths, on="School Name").merge(passing_reading,on="School Name").merge(overall_passing,on="School Name")

# School summary results calculation
school_summary["% Passing Maths"] = (school_summary["% Passing Maths"]/school_summary["Total Students"])*100
school_summary["% Passing Reading"] = (school_summary["% Passing Reading"]/school_summary["Total Students"])*100
school_summary["% Overall Passing"] = (school_summary["% Overall Passing"]/school_summary["Total Students"])*100

school_summary["Total School Budget"] = school_summary["Total School Budget"].astype(float).map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].astype(float).map("${:,.2f}".format)

# Schools in alphabetical order
school_summary = school_summary.sort_values("School Name")

# Set the School Name as the table index
school_summary.set_index("School Name")

# export school_summary to csv
#school_summary.to_csv('school_summary.csv', index=False)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.
#Create a pandas series for each year. Hint: use a conditional statement.
#Group each series by school

#Year 9s
year_9 = pd.DataFrame(school_data_complete.loc[school_data_complete["year"] == 9,
                                                                  ["School Name", "Maths Score"]]
                    ).groupby("School Name").mean()
year_9 = year_9.rename(columns = {"Maths Score":"Year 9"})
year_9 = year_9[["Year 9"]].sort_values("Year 9", ascending=False)

#Year 10s
year_10 = pd.DataFrame(school_data_complete.loc[school_data_complete["year"] == 10,
                                                                   ["School Name", "Maths Score"]]
                     ).groupby("School Name").mean()
year_10 = year_10.rename(columns = {"Maths Score":"Year 10"})
year_10 = year_10[["Year 10"]].sort_values("Year 10", ascending=False)

#Year 11s
year_11 = pd.DataFrame(school_data_complete.loc[school_data_complete["year"] == 11,
                                                                    ["School Name", "Maths Score"]]
                     ).groupby("School Name").mean()
year_11["Year 11"] = year_11.rename(columns = {"Maths Score":"Year 11"})
year_11 = year_11[["Year 11"]].sort_values("Year 11", ascending=False)

#Year 12s
year_12 = pd.DataFrame(school_data_complete.loc[school_data_complete["year"] == 12,
                                                                   ["School Name", "Maths Score"]]
                      ).groupby("School Name").mean()
year_12["Year 12"] = year_12.rename(columns = {"Maths Score":"Year 12"})
year_12 = year_12[["Year 12"]].sort_values("Year 12", ascending=False)

# Maths Scores by Year (MSBY)
#Combine the series into a dataframe
MSBY = year_9.merge(year_10,on="School Name").merge(year_11,on="School Name").merge(year_12,on="School Name")
MSBY.sort_values("School Name").head()

# export MSBY to csv
#MSBY.to_csv('MSBY.csv', index=False)
```
